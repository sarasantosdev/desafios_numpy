### Calculando o coeficiente angular e linear das laranjas

```python
X = orange_diam
Y = orange_peso
n = np.size(Y)
```

```python
def coef_angular(X,Y,n):
    calculo = (n * (np.sum(X * Y)) - np.sum(X) * np.sum(Y))/(n * (np.sum(X**2)) - (np.sum(X)**2))
    return calculo
```

```python
def coef_linear(X,Y,a):
    calculo = np.mean(Y) - a * np.mean(X)
    return calculo
```

```python
coef_angular(X,Y,n)
```

```python
coef_linear(X,Y,a)
```
